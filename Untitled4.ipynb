```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense,GlobalAveragePooling2D,Dropout
import cv2
from PIL import Image
import requests
from io import BytesIO
```

```python
path = "/content/DATA/"
```

```python
h, w = 299, 299
x = []
y = []
images = []
d = {}
for i, f_name in enumerate(os.listdir(path)):
  d [i] = f_name
  for j in os.listdir(path + f_name):
    y.append(i)
    img = cv2.imread(path + f_name + "/" + j)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (h, w))
    img_p = tf.keras.applications.inception_resnet_v2.preprocess_input(img)
    x.append(img_p)
    images.append(img)
x = np.array(x)
y = np.array(y)
images = np.array(images)
print(d)
images.shape, x.shape, y.shape
```

```python
plt.figure(figsize=(15, 25))
for i in range(100):
  ind = np.random.randint(0, 660)
  plt.subplot(10, 10, i+1)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(images[ind])
  plt.title(d[y[ind]])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
y_train_cat = tf.keras.utils.to_categorical(y_train)
y_test_cat = tf.keras.utils.to_categorical(y_test)
```

```python
inc_res = tf.keras.applications.InceptionResNetV2(include_top=False)
```

```python
for i in inc_res.layers:
    i.trainable = False
```

```python
inp = Input(shape=x_train.shape[1:])
m = inc_res(inp)
g = GlobalAveragePooling2D()(m)
d_0 = Dense(150, activation="relu")(g)
drop = Dropout(0.3)(d_0)
out = Dense(11, activation="softmax")(drop)

model = tf.keras.Model(inp, out)
model.summary()
```

```python
model.compile(optimizer= "adam", loss = "categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(x_train, y_train_cat, batch_size=10, epochs=3, validation_data=(x_test, y_test_cat))
```

```python
response = requests.get("https://warehouse1.indicia.org.uk/upload/o_1hv2t6bb8e1iv7q11t51mh91pn14k.jpg")
img = Image.open(BytesIO(response.content))

img = img.resize((w, h)).convert("RGB")
img = np.array(img)
plt.imshow(img)
print(img.shape)

img = tf.keras.applications.inception_resnet_v2.preprocess_input(img)
img = np.expand_dims(img, axis = 0)
p = model.predict(img)
p_ind = p.argsort()[0][-3:]
p_val = p[0][p_ind]
for i, j in zip(p_ind, p_val):
    print(d[i], "-", round(j, 4))
```
